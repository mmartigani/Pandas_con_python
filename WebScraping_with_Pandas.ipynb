```python
df=pd.read_csv('players_20.csv')
#seleccionar el index 
df.set_index('short_name', inplace=True)
#seleccionamos las columnas 
df=df[['long_name','age','dob','height_cm','weight_kg','nationality', 'club']]
df
```

```python
df.loc['L. Messi',]
```

```python
df.loc['L. Messi','height_cm']
```

```python
df.loc['Cristiano Ronaldo','weight_kg']
```

```python
#obtenemos todas las filas de las columna height_cm
df.loc[:,'height_cm']
```

```python
#todas las columnas de una fila 
df.loc['L. Messi',:]
```
