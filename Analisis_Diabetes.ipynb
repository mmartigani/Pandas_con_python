```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline 
```

```python
df=pd.read_csv('diabetes.csv')
df
```

```python
df['Pregnancies/BloodPressure']=(df['Pregnancies']/df['BloodPressure']).astype(float)
df
```

```python
Total_Pregnancies=df['Pregnancies'].sum()
Total_Pregnancies
```

```python
promedio_age=df['Age'].mean()
promedio_age
```

```python
plt.figure(figsize=(13,6))
age=df['Age'].sum()-Total_Pregnancies
data=[age, Total_Pregnancies]
labels=['embarazos','edades']
color=sns.color_palette('Set2')
plt.pie(data, labels=labels , colors=color,  autopct='%.0f%%')
plt.show()

```

```python
df['Age'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
df['Age'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
debajo_20=df[df['Age']<=20]
entre_20_25=df[(df['Age'] >20)&(df['Age'] <=25)]
entre_25_30=df[df['Age']<=25 & (df['Age'] <=30)]
arriba_30=df[df['Age']>30]
```

```python
x=np.array([debajo_20['Age'].count(),entre_20_25['Age'].count(),entre_25_30['Age'].count(),arriba_30['Age'].count()])
my_labels=['<=20','>20 &  <=25','<=25 &  <=30 ','>30 ']
plt.title('total con las mismas edades')
plt.pie(x, labels= my_labels,autopct='%.1f%%')
plt.show()
```
